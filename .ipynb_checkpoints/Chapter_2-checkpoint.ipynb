##Visualization

```python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
from sklearn.datasets import load_iris
from itertools import combinations
```

```python
data = load_iris()

features = data.data
feature_names = data.feature_names
target = data.target
target_names = data.target_names

def get_combos(li, r):
    return [c for c in combinations(li, r)]

def plot_features(x, y):
    for t in range(3):
        if t==0:
            c = 'r'
            marker = '>'
        elif t==1:
            c = 'g'
            marker = 'o'
        elif t==2:
            c = 'b'
            marker = 'x'
        plt.scatter(features[target==t,x],
                    features[target==t,y],
                    marker=marker,
                    c=c)
        plt.gca().set_xlabel(feature_names[x])
        plt.gca().set_ylabel(feature_names[y])

for x,y in get_combos(range(4), 2):
    plt.figure()
    plot_features(x,y)
```

petal length allows us to separate out iris setosa.

```python
labels = target_names[target] #array of strings
plength = features[:, 2]
is_setosa = (labels=='setosa')

#important step:
max_setosa = plength[is_setosa].max()
min_non_setosa = plength[~is_setosa].min()
print('Maximum of setosa: {0}.'.format(max_setosa))
print('Minimum of others: {0}.'.format(min_non_setosa))
```

Therefore, if petal length is smaller than 2, we can conclude this is an Iris Setosa.

We can only look for "best possible" separation between other species.

```python
features =  features[~is_setosa]
labels = labels[~is_setosa]
is_virginica = (labels == 'virginica')
is_virginica[:5]
```

```python
best_acc = -1.0
for fi in range(features.shape[1]):
    # We are going to test all possible thresholds
    thresh = features[:, fi]
    for t in thresh:
        feature_i =  features[:, fi]
        pred = (feature_i > t) #viginica is greater
        acc = (pred == is_virginica).mean()
        rev_acc = (pred == ~is_virginica).mean()
        if rev_acc > acc:
            reverse = True
            acc = rev_acc
        else:
            reverse = False
        
        if acc > best_acc:
            best_acc = acc
            best_fi = fi
            best_t = t
            best_reverse = reverse
```

```python
def is_virginica(fi, t, reverse, example):
    "Apply threshold model to a new example"
    test = example[fi] > t
    if reverse:
        test = not test
    return test
```

###cross validation
Ideally we'd like to use all of the data for training and testing, but we can't.  Leave out a subset of the training set.  Train on the rest of the set and test on this subset.  Repeat for all elements in the dataset.

This is computationally expensive though.  **x-fold cross validation** attempts to solve this.  Instead of leaving one out, leave out say, 20%, in a 5 fold cross validation.

With less data, use up to 10 or 20 folds.

##More Complex Classifiers
What makes up a classification model?
* **The Structure of the Model** - how does it make decisions?
* **The search procedure** - How do we find the model we need to use?
* **gain or loss function** - Which possibilities tested should be returned?  Optimize so that model makes fewer erros of a specific kind.

**feature engineering** - derive new combined features.  Should vary with what matters but be invariant with what does not.

**feature selection** is choosing good features automatically.
